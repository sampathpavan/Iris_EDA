# Some mathematical operations 

```python
# MEAN
```

```python
sepalLength_mean=round(iris['SepalLengthCm'].mean(),1)
print("Mean of sepalLength is ",sepalLength_mean)

sepalWidth_mean=round(iris['SepalWidthCm'].mean(),1)
print("Mean of sepalWidth is ",sepalWidth_mean)

petalLength_mean=round(iris['PetalLengthCm'].mean(),1)
print("Mean of petalLength is ",petalLength_mean)

petalWidth_mean=round(iris['PetalWidthCm'].mean(),1)
print("Mean of petalWidth is ",petalWidth_mean)

```

```python
# MEDIAN
```

```python
print('Median for SepalLengthCm:',iris['SepalLengthCm'].median())

print('Median for SepalWidthCm:',iris['SepalWidthCm'].median())

print('Median for PetalLengthCm:',iris['PetalLengthCm'].median())

print('Median for PetalWidthCm:',iris['PetalWidthCm'].median())
```

```python
# MODE
```

```python
print('Mode for SepalLengthCm:',iris['SepalLengthCm'].mode())

print('Mode for SepalWidthCm:',iris['SepalWidthCm'].mode())

print('Mode for PetalLengthCm:',iris['PetalLengthCm'].mode())

print('Mode for PetalWidthCm:',iris['PetalWidthCm'].mode())
```

# Probability of unique values in each column

```python
sepalLength_unique=iris['SepalLengthCm'].unique()
print(f'Total distinct elements in SepalLengthCm are {len(sepalLength_unique)} there probabilities are: ')
for i in sepalLength_unique:
    count = (iris['SepalLengthCm']==i).sum()    
    print(f'P({i}) = {round((count/iris.shape[0]),2)}')
```

```python
sepalWidth_unique=iris['SepalWidthCm'].unique()
print(f'Total distinct elements in SepalWidthCm are {len(sepalWidth_unique)} there probabilities are: ')
for i in sepalWidth_unique:
    count = (iris['SepalWidthCm']==i).sum()    
    print(f'P({i}) = {round((count/iris.shape[0]),2)}')
```

```python
petalLength_unique=iris['PetalLengthCm'].unique()
print(f'Total distinct elements in PetalLengthCm are {len(petalLength_unique)} there probabilities are: ')
for i in petalLength_unique:
    count = (iris['PetalLengthCm']==i).sum()    
    print(f'P({i}) = {round((count/iris.shape[0]),2)}')
```

```python
petalWidth_unique=iris['PetalWidthCm'].unique()
print(f'Total distinct elements in PetalWidthCm are {len(petalWidth_unique)} there probabilities are: ')
for i in petalWidth_unique:
    count = (iris['PetalWidthCm']==i).sum()    
    print(f'P({i}) = {round((count/iris.shape[0]),2)}')
```
